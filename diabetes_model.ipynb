```python
#Tentukan library yang digunakan

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#load dataset
diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
#memisahkan data dan label pisahkan diagnosa(pregencies dan ll ) dengan atribut (outcome)
X = diabetes_dataset.drop (columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
# jadi kita berhasil pisahkan diagnosa nya dan labelnya atributnya
print(Y)
```

```python
#pisahkan data training dan data testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify=Y, random_state=2)
```

```python
#kita cetak nilai yang di training dan di testing
print(X.shape, X_train.shape, X_test.shape)
#data dan atribut karena kurangi 1 untuk label
```

```python
# Membuat data latih menggunakan algoritma SVM
classifier = svm.SVC(kernel='linear')
```

```python
#yang diambil  x training y labelnya
classifier.fit(X_train, Y_train)
#sudah memasukkan data training dan testing
#sebelum membuat prediksi kita testing dulu
```

```python
# Membuat model evaluasi untuk mengukur tingkat akurasi
# data trining yang diambil adalah x training
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
#perintah ini adalah deteksi akurasi untuk  data training
```

```python
print('Akurasi data training adalah =', training_data_accuracy)
```

```python
# Membuat Model Prediksi
input_data = (6.0, 148.0, 72.0, 35.0, 169.0, 33.6, 0.627, 50.0, 5.135798)
input_data_as_numpy_array = np.array(input_data)
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)
std_data = input_data_reshape
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('pasien tidak terkena diabetes')
else :
  print('Pasien terkena diabetes')
```

```python
# Membuat Model Prediksi
input_data = (1.0, 85.0, 66.0, 29.0, 58.6, 26.6, 0.351, 31.0, 4.087656)
input_data_as_numpy_array = np.array(input_data)
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)
std_data = input_data_reshape
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('pasien tidak terkena diabetes')
else :
  print('Pasien terkena diabetes')
```
